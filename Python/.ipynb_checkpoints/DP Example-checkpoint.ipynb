### Variations of Fibonacci

#### Recursive
Time complexity: 
$O(2^n)$

```python
def fib(n):
    if n <= 1:
        return 1
    else:
        f = fib(n-1) + fib(n-2)
    return f
```

#### Top-down DP
Time complexity: $O(n)$

```python
memo = [0 for i in range(1000)]
def fib_memo(n):
    if n <= 1:
        return 1
    if memo[n] != 0:
        return memo[n]
    else:
        f = fib_memo(n-1) + fib_memo(n-2)
        memo[n] = f
    return memo[n]
```

#### Bottom-up DP
Time complexity: $O(n)$

```python
memo = [0 for i in range(1000)]
def fib_btm_up(n):
    for k in range(0, n+1):
        if k <= 1: 
            f = 1
        else:
            f = memo[k-1] + memo[k-2]
        memo[k] = f
    return memo[n]
```

### Run tests

```python
time_scale = []
total = 10

for n in range(0, total):
    st = time.time()
    print(fib(n), end=' ') # Running for Recursive case
    et = time.time()
    time_scale.append((et - st))
```

```python
time_scale_2 = []

for n in range(0, total):
    st = time.time()
    print(fib_memo(n), end=' ') # Running for TD DP
    et = time.time()
    time_scale_2.append((et - st))
```

```python
time_scale_3 = []

for n in range(0, total):
    st = time.time()
    print(fib_btm_up(n), end=' ') # Running for BU DP
    et = time.time()
    time_scale_3.append((et - st))
```

```python
plt.figure(figsize=[10, 7])
plt.plot(time_scale, marker='o', color='b', label='Recursive Code')
plt.plot(time_scale_2, marker='o', color='r', label='DP Memoization')
plt.plot(time_scale_3, marker='o', color='g', label='DP Bottom-Up')
plt.xlabel('n')
plt.ylabel('time in microseconds')
plt.legend()
plt.show()
```

### Rod-Cutting
Time complexity: 
$\sum_{i=1}^{n+1} \sum_{j=1}^{i} (1+3) = O(n^2)$

```python
def mem_rod_cutting(prices, n):
    mem = {0:0}
    for j in range(1, n+1):
        q = float('-inf')
        for i in range(j):
            q = max(q, p[i] + mem[j-i-1])
        mem[j] = q
    return mem
```

### Coin Change
Time complexity: 
$\sum_{i=1}^{n+1} \sum_{j=0}^{m} (2+3+1+1+1) = O(n*m)$

```python
def coin_change(denom, m, n):
    mem = {0:0}
    for j in range(1, n+1): # n
        min_ = float('inf')
        for i in range(0, m): # m
            if denom[i] <= j:
                min_ = min(1 + mem[j - denom[i]], min_)
        mem[j] = min_
    return mem
```

```python
denom = [1,3,4,6,5]
n = 13
m = len(denom)
mem = coin_change(denom, m=m, n=n)
print(f"{mem}\nMinimum coins required to make change of {n} is {mem[n]}")
```

### 0/1 Knapsack problem
Time complexity: 
$\sum_{i=0}^{n} \sum_{j=0}^{W} (4+1+3+2+7) = O(n*W)$

```python
def knapsack(wt, p, n, W):
    mem = np.zeros((n+1, W+1))
    for i in range(n+1):
        for w in range(W+1):
            if i == 0 or w == 0:
                mem[i][w] = 0
            elif wt[i-1] > w:
                mem[i][w] = mem[i-1][w]
            else:
                mem[i][w] = max(p[i-1] + mem[i-1][w-wt[i-1]], mem[i-1][w])
    return mem
```

```python
wt = [1,2,5,6,7,9]
p = [2,7,18,23,29,35]
n = len(wt) # or len(p)
W = 12
mem = knapsack(wt, p, n, W)
print(f"{mem}\nMaximum profit for the knapsack of weight {W}: {mem[n][W]}")
```
